## Missing data

```python
## Missing data
features_with_NAN = [features for features in data.columns if data[features].isnull().sum() >=1 ]
#features with NAN Value
features_with_NAN
```

```python
## Percentage of the nan value
for features in features_with_NAN:
    print(features, np.round(data[features].isnull().mean()*100, 4),"% of missing value")
```

```python
#take each and every feature if NAN -> 1, else -> 0 
# #group by 1 and 0

for features in features_with_NAN:
    #copy the dataset :Shallow copy()
    df = data.copy()
    #lets make variable that indicate 1 and 0
    data[features] = np.where(data[features].isnull(),1,0)
    #lets calculate the mean SalePricew where information is missing
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.title(features)
    plt.show()
```
